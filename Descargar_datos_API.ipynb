```python
import urllib
import urllib.request
import pandas as pd
import requests
```

```python
url = 'https://opendata-ajuntament.barcelona.cat/data/api/action/datastore_search?resource_id=9caf3470-f325-4079-9884-8bc819104884'

# Hacer la solicitud GET
response = requests.get(url)
if response.status_code == 200:
    # Obtener los datos en formato JSON
    datos = response.json()
    
    # Extraer los registros (el contenido que quieres convertir a un DataFrame)
    registros = datos['result']['records']
    
    # Crear el DataFrame a partir de los registros
    df = pd.DataFrame(registros)
    print(df)
```

##### Crear una API para que puedan acceder a nuestros datos

```python
#pip install Flask
```

```python
from flask import Flask, jsonify
import json
```

```python
app = Flask(__name__)
```

```python
# Cargar los datos desde el archivo JSON
def cargar_datos():
    with open('fires_mercats_final_records.json') as archivo:
        datos = json.load(archivo)
    return datos
```

```python
# Ruta para obtener todos los datos
@app.route('/api/datos', methods=['GET'])
def obtener_datos():
    datos = cargar_datos()  # Cargar los datos del archivo JSON
    return jsonify(datos)    # Devolver los datos en formato JSON
```
